```python
import pandas as pd
```

```python
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')
reavealed_targets = pd.read_csv('revealed_targets.csv')
sample_submission = pd.read_csv('sample_submission.csv')
```

```python
reavealed_targets = reavealed_targets.dropna()
reavealed_targets['date_id'] = reavealed_targets['date_id'].astype(int)
reavealed_targets['seconds_in_bucket'] = reavealed_targets['seconds_in_bucket'].astype(int)
reavealed_targets['stock_id'] = reavealed_targets['stock_id'].astype(int)

reavealed_targets['row_id'] = reavealed_targets['date_id'].astype(str) + '_' + reavealed_targets['seconds_in_bucket'].astype(str) + '_' +  reavealed_targets['stock_id'].astype(str)
```

```python
reavealed_targets
```

```python
merged_df = pd.merge(test_data, reavealed_targets, on='row_id', how='inner')

merged_df = merged_df.drop(['stock_id_y', 'date_id_y', 'seconds_in_bucket_y', 'time_id_y', 'revealed_date_id', 'revealed_time_id'], axis=1)

merged_df
```

```python
train_data['far_price'].fillna(train_data['far_price'].mean(), inplace=True)
train_data['near_price'].fillna(train_data['near_price'].mean(), inplace=True)
train_data['reference_price'].fillna(train_data['reference_price'].mean(), inplace=True)
train_data['imbalance_size'].fillna(train_data['imbalance_size'].mean(), inplace=True)
train_data['matched_size'].fillna(train_data['matched_size'].mean(), inplace=True)
train_data['wap'].fillna(train_data['wap'].mean(), inplace=True)
train_data['target'].fillna(train_data['target'].mean(), inplace=True)
train_data['bid_price'].fillna(train_data['bid_price'].mean(), inplace=True)
train_data['ask_price'].fillna(train_data['ask_price'].mean(), inplace=True)
```

```python
merged_df['far_price'].fillna(merged_df['far_price'].mean(), inplace=True)
merged_df['near_price'].fillna(merged_df['near_price'].mean(), inplace=True)
merged_df['reference_price'].fillna(merged_df['reference_price'].mean(), inplace=True)
merged_df['imbalance_size'].fillna(merged_df['imbalance_size'].mean(), inplace=True)
merged_df['matched_size'].fillna(merged_df['matched_size'].mean(), inplace=True)
merged_df['wap'].fillna(merged_df['wap'].mean(), inplace=True)
merged_df['revealed_target'].fillna(merged_df['revealed_target'].mean(), inplace=True)
merged_df['bid_price'].fillna(merged_df['bid_price'].mean(), inplace=True)
merged_df['ask_price'].fillna(merged_df['ask_price'].mean(), inplace=True)
```

```python
train_data
```

```python
import numpy as np
train_data['log_imbalance_size'] = np.log1p(train_data['imbalance_size'])
merged_df['log_imbalance_size'] = np.log1p(merged_df['imbalance_size'])

train_data['imbalance_size_squared'] = train_data['imbalance_size'] ** 2
train_data['sqrt_imbalance_size'] = np.sqrt(train_data['imbalance_size'])
# train_data['exp_imbalance_size'] = np.exp(train_data['imbalance_size'])

merged_df['imbalance_size_squared'] = merged_df['imbalance_size'] ** 2
merged_df['sqrt_imbalance_size'] = np.sqrt(merged_df['imbalance_size'])
# merged_df['exp_imbalance_size'] = np.exp(merged_df['imbalance_size'])


train_data['imbalance_ratio'] = train_data['imbalance_size'] / train_data['matched_size']
train_data['imbl_size1'] = (train_data['bid_size']-train_data['ask_size']) / (train_data['bid_size']+train_data['ask_size'])
train_data['imbl_size2'] = (train_data['imbalance_size']-train_data['matched_size']) / (train_data['imbalance_size']+train_data['matched_size'])


merged_df['imbalance_ratio'] = merged_df['imbalance_size'] / merged_df['matched_size']
merged_df['imbl_size1'] = (merged_df['bid_size']-merged_df['ask_size']) / (merged_df['bid_size']+merged_df['ask_size'])
merged_df['imbl_size2'] = (merged_df['imbalance_size']-merged_df['matched_size']) / (merged_df['imbalance_size']+merged_df['matched_size'])
```

```python
train_data = train_data.dropna()

# X_train = train_data.drop(['stock_id', 'date_id', 'time_id', 'target', 'row_id'], axis=1)
X_train = train_data.drop(['target', 'row_id'], axis=1)
y_train = train_data['target']
# X_test = merged_df.drop(['stock_id_x', 'date_id_x', 'time_id_x', 'row_id', 'revealed_target'], axis=1)
X_test = merged_df.drop(['revealed_target', 'row_id'], axis=1)
y_test = merged_df['revealed_target']
X_test.rename(columns={'seconds_in_bucket_x': 'seconds_in_bucket'}, inplace=True)
X_test.rename(columns={'stock_id_x': 'stock_id'}, inplace=True)
X_test.rename(columns={'date_id_x': 'date_id'}, inplace=True)
X_test.rename(columns={'time_id_x': 'time_id'}, inplace=True)

X_train = pd.get_dummies(X_train, columns=['imbalance_buy_sell_flag'], drop_first=False)
X_test = pd.get_dummies(X_test, columns=['imbalance_buy_sell_flag'], drop_first=False)
```
